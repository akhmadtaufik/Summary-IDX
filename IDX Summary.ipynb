```python
old = old.reset_index()
for i in ['Open', 'High', 'Close', 'Low']:
  old[i] = old[i].astype('float64')
```

```python
import plotly.graph_objects as go 
fig = go.Figure(data=[go.Candlestick(x=old['Date'], 
                                     open=old['Open'],
                                     high=old['High'],
                                     low=old['Low'],
                                     close=old['Close'])])
fig.show()
```

```python
import plotly.express as px

fig = px.line(old, x='Date', y='Open', title='Bank BCA Stock Prices')
fig.show()
```

```python
# OHLC Charts
fig = go.Figure(data=go.Ohlc(x=old['Date'],
                    open=old['Open'],
                    high=old['High'],
                    low=old['Low'],
                    close=old['Close']))
fig.show()
```
